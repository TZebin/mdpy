```python
import numpy as np
from numpy.linalg import pinv
```

```python
np.set_printoptions(precision=6, suppress=True)
```

```python
import mdpy
```

```python
ns = 5

I = np.eye(ns)
P = mdpy.rand_p(ns)
```

```python
d = mdpy.stationary(P)
print(d)
```

```python
gm = 0.9
```

```python
lst = []
for lm in np.linspace(0, 1, 11):
    lst.append(pinv(I - lm*gm*P) @ (I - gm*P))
```

```python
for i in lst: print(i, end="\n\n")
```
